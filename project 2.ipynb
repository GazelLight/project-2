### filling in the blanks

```python
cars_df['model_year'] = cars_df.groupby(['model'])['model_year'].transform(lambda grp: grp.fillna(np.mean(grp)))
cars_df.info()
```

```python
cars_df.head(10)
```

i'll use the floor method on the columns so ther will be no decimal numbers.

```python
cars_df["model_year"] = cars_df['model_year'].apply(np.floor)
```

```python
cars_df['cylinders'] = cars_df.groupby(['model'])['cylinders'].transform(lambda grp: grp.fillna(np.mean(grp)))
cars_df["cylinders"] = cars_df['cylinders'].apply(np.floor)
cars_df.info()
```

```python
cars_df['odometer'] = cars_df.groupby(['model'])['odometer'].transform(lambda grp: grp.fillna(np.mean(grp)))
cars_df['odometer'] = cars_df.groupby(['model_year'])['odometer'].transform(lambda grp: grp.fillna(np.mean(grp)))
cars_df["odometer"] = cars_df['odometer'].apply(np.floor)
cars_df.info()
```

in the odometer column, I used another transform using the model year column, becouse using only the model, did not fill all the missing values. 


in the paint_color column, i can't fill in the missing values. that is because the paint of the car can not be determined by any other feature of the car. so trying to fill the values in the column, can harm the the data and mislead my conclusion, regarding the colors affect on the price of the car.

### categorisation of condition

to make the condition column more comfortable to use, i'm going to change the string categorias in the column to integers. to do that i created a function that does the categorisation, and applied it on the column 

```python
def num_condition(row):
    
    num = row["condition"]
    
    if num == "new":
        return 5
    if num == "like new":
        return 4
    if num == "excellent":
        return 3
    if num == "good":
        return 2
    if num == "fair":
        return 1
    if num == "salvage":
        return 0
    else:
        return num
```

```python
cars_df["condition"] = cars_df.apply(num_condition, axis=1)
```

```python
def boxplot_cons(column):
    Q1 = cars_df[column].quantile(0.25)
    Q3 = cars_df[column].quantile(0.75)
    IQR = Q3 - Q1
    cars_df.boxplot(column)
    plt.hlines(y = Q1-1.5*IQR, xmin = 0.9, xmax = 1.1, color = "red")
    plt.hlines(y = Q3 + 1.5*IQR, xmin = 0.9, xmax = 1.1, color = "red")
```
